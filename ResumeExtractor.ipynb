```python
from pdfminer.high_level import extract_text
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
def extract_resume_text(file):

    text = extract_text(file)

    return text
```

```python
def preprocess(text):

    text = text.lower()

    tokens = word_tokenize(text)

    stop_words = set(stopwords.words('english'))

    tokens = [w for w in tokens if w not in stop_words]

    tokens = [w for w in tokens if w not in string.punctuation]

    return tokens
```

```python
skills_db = [

'python',
'sql',
'machine learning',
'deep learning',
'pandas',
'numpy',
'power bi',
'tableau',
'statistics',
'data analysis',
'excel',
'nlp',
'tensorflow',
'scikit-learn'

]
```

```python
def extract_skills(text):

    text = text.lower()

    skills_found = []

    for skill in skills_db:

        if skill in text:
            skills_found.append(skill)

    return set(skills_found)
```

```python
job_description = """
Looking for a data scientist with skills in Python,
SQL, Machine Learning, Deep Learning, and Statistics.
"""
```

```python
job_skills = extract_skills(job_description)

print("Job Skills Required:", job_skills)
```

```python
resumes = [

"resume_data_scientist.pdf",
"resume_data_analyst.pdf",
"resume_ml_engineer.pdf"

]
```

```python
scores = {}

for resume in resumes:

    text = extract_resume_text(resume)

    resume_skills = extract_skills(text)

    matched = resume_skills.intersection(job_skills)

    score = (len(matched) / len(job_skills)) * 100

    scores[resume] = score

    print("\nResume:", resume)

    print("Skills:", resume_skills)

    print("Matched Skills:", matched)

    print("Score:", score)
```
